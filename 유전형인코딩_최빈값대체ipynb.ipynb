```python

# üìå 2. Îç∞Ïù¥ÌÑ∞ Î°úÎî©
import pandas as pd
import numpy as np

file_name = "ÌÜ†ÎßàÌÜ†_Ïú†Ï†ÑÌòï.CSV"  # ÏóÖÎ°úÎìúÌïú ÌååÏùºÎ™ÖÏóê ÎßûÍ≤å ÏàòÏ†ï
geno_raw = pd.read_csv(file_name, index_col=0)

# üìå 3. Ï†ÑÏπò Î∞è 'Chr', 'Position (bp)' Ï†úÍ±∞
geno_T = geno_raw.drop(columns=["Chr", "Position (bp)"]).T.replace("NN", np.nan)

# üìå 4. ÏóºÍ∏∞ Ïù∏ÏΩîÎî© Í∑úÏπô (AC/CA ‚Üí 12 ÌòïÌÉú)
base_encoding = {'A': '1', 'C': '2', 'G': '3', 'T': '4'}
def encode_snp_sorted(snp):
    if pd.isna(snp):
        return np.nan
    sorted_snp = ''.join(sorted(snp))
    return int(''.join([base_encoding.get(base, '0') for base in sorted_snp]))

# üìå 5. Î¨∏Ïûê SNP ‚Üí Ïà´Ïûê Ïù∏ÏΩîÎî©
geno_encoded = geno_T.applymap(encode_snp_sorted)

# üìå 6. Í≤∞Ï∏°Ïπò ‚Üí ÏµúÎπàÍ∞íÏúºÎ°ú ÎåÄÏ≤¥ + Ï†ïÏàòÌòï Î≥ÄÌôò
geno_filled = geno_encoded.apply(
    lambda col: col.fillna(col.mode().iloc[0]).astype(int), axis=0
)

# üìå 7. Ï†ÑÏπò (ÏÉòÌîåÏùÑ Ìñâ, SNPÎ•º Ïó¥Î°ú Î≥¥Í∏∞ Ï¢ãÍ≤å)
geno_final = geno_filled.T

# üìå 8. CSVÎ°ú Ï†ÄÏû•
output_filename = "Ï†ÑÏ≤òÎ¶¨Îêú_Ïú†Ï†ÑÌòï_ÏµúÎπàÍ∞í_Ï†ÑÏπò.csv"
geno_final.to_csv(output_filename, encoding='utf-8-sig', index=True)

# üìå 9. Îã§Ïö¥Î°úÎìú
files.download(output_filename)
```
